```python
pacs = 0
data = data[["ArticleID", "PACS", "FullName", "AuthorID"]]

data_i = data[data["PACS"] == pacs]
data_i

data_i = data_i[["ArticleID", "AuthorID"]]
data_i
```

```python
data_grouped = data_i.groupby('ArticleID')['AuthorID'].apply(list).to_frame()
data_grouped
```

```python
edges = data_grouped.to_dict()
edges = edges['AuthorID']
```

```python
H = xgi.Hypergraph(edges)
```

```python
print(H)
```

```python
xgi.write_json(H, f"../data/PACS{pacs}.json")
```

## loop over all PACS:

```python
for pacs in range(10):
    
    data = data[["ArticleID", "PACS", "FullName", "AuthorID"]]
    data_i = data[data["PACS"] == pacs]
    data_i = data_i[["ArticleID", "AuthorID"]]
    
    data_grouped = data_i.groupby('ArticleID')['AuthorID'].apply(list).to_frame()
    edges = data_grouped.to_dict()
    edges = edges['AuthorID']
    
    H = xgi.Hypergraph(edges)
    xgi.write_json(H, f"../data/PACS{pacs}.json")
```
